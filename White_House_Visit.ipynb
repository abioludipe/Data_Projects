# WHITE HOUSE VISIT ANALYSIS

**Next we need to determine how long the meetings take in each venue.** 

- From the data we can see that the numbers were increasing steadily till it got to 2265 and then it jumped to 11605. So it is safe to assume that most people visiting the white house do so for occasions. In analysing the amount of time the meetings take we will not be considering meetings held in the State floor and South lawn. Let's create a duplicate of our data and drop the rows where meeting room is state floor or south lawn.

```python
wh_copy = wh_visit.copy(deep = True)
```

```python
wh_copy = wh_copy[wh_copy.meeting_room != "state floor"]
```

```python
wh_copy = wh_copy[wh_copy.meeting_room != "south lawn"]
```

```python
number_of_visits = wh_copy["meeting_room"].value_counts().sort_values(ascending = False).head(5)
number_of_visits
```

```python
wh_copy["appt_start_date"] = pd.to_datetime(wh_copy["appt_start_date"],format="%m/%d/%y %H:%M")
# wh_copy["appt_start_date"] = wh_copy["appt_start_date"].dt.strftime("%d/%m/%Y %H")
# wh_copy["appt_start_date"] = wh_copy["appt_start_date"].astype("datetime64[ns]")
```

```python
wh_copy["appt_end_date"] = pd.to_datetime(wh_copy["appt_end_date"],format="%m/%d/%y %H:%M")
# wh_copy["appt_end_date"] = wh_copy["appt_end_date"].dt.strftime("%d/%m/%Y %H")
# wh_copy["appt_end_date"] = wh_copy["appt_end_date"].astype("datetime64[ns]")
```

```python
wh_copy.head(10)
```

```python
wh_copy["t_diff"] = wh_copy["appt_end_date"] - wh_copy["appt_start_date"]
```

```python
wh_copy.head(10)
```

```python
meeting_room = wh_copy["meeting_room"].value_counts()
first_5_room = meeting_room[meeting_room > 550].index
first_5_room
```

```python
meeting_room_sum = {}
meeting_room_mean = {}

for room in first_5_room:
    room_only = wh_copy[wh_copy["meeting_room"] == room]
    room_only_sum = room_only["t_diff"].sum()
    room_only_mean = room_only["t_diff"].mean()
    meeting_room_sum[room] = room_only_sum
    meeting_room_mean[room] = room_only_mean
    
meeting_room_sum
meeting_room_mean
```

```python
meeting_hours = pd.Series(meeting_room_sum)
meeting_hours_avg = pd.Series(meeting_room_mean)
```

```python
meeting_hours_per_room = pd.DataFrame(meeting_hours, columns = ["time_spent_per_room"])
meeting_hours_per_room["avg_time_spent"] = meeting_hours_avg
meeting_hours_per_room["Number_of_visits"] = number_of_visits
meeting_hours_per_room
```

From the above the data reads from the 6th of January 2015 to 18th of December 2015. The Oval Office was visited the most also is the room that has the highest total sum of hours spent. However, the average number of hours spent in a meeting in the Oval Office is less than the East wing and East room. 

- *The East Wing is a part of the White House Complex. It is a two-story structure east of the White House Executive Residence, the home of the president of the United States. While the West Wing generally serves the president's executive office staff, the East Wing serves as office space for the first lady and her staff, including the White House social secretary, White House Graphics and Calligraphy Office and correspondence staff. The East Wing also includes the visitors' entrance, and the East Colonnade, a corridor connecting the body of the East Wing to the residence. Along the corridor is the White House theater, also called the Family theater. Social and touring visitors to the White House usually enter through the East Wing.*


- *The East Room is an event and reception room in the White House, the home of the President of the United States. The East Room is the largest room in the Executive Mansion; it is used for dances, receptions, press conferences, ceremonies, concerts, and banquets.*


**Since the East Room is the Residential arm of the White House, this explains why less people visit the but spend more time since it is the room with highest mean hours.**

**The East wing is used for receptions and ceremonies, this gatherings usually take longer than short meetings but don't occur often hence it has an high average time with less visit.**
