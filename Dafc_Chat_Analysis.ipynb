# Dafc Chat Analysis

### The objective is to find out the most active members of the group page and as well identify the inactive members and be able to reach out to them

```python
#importing libraties

import pandas as pd
import numpy as np
import re
from csv import reader
```

```python
#opening the file into a list

open_chat = open('Dafc_chat.txt', encoding = 'utf8')
read_chat = reader(open_chat)
dafc = list(read_chat)
```

```python
#exploring the dataset

def explore_data(dataset, start, end, row_column = False):
    dataslice = dataset[start:end]
    for row in dataslice:
        print(row)
        print('\n')
    if row_column:
        print('Number of rows: ',  len(dataset))
        print('Number of columns ',  len(dataset[0]))
        print('\n')
```

```python
explore_data(dafc, 0,5,True)
```

```python
#reading the list into a pandas dataframe

dafc = pd.DataFrame(dafc)
```

```python
#peeping into the df

dafc.head()
```

```python
#dropping columns with less than 100 non null elements

dafc = dafc.drop(labels = range(2,38), axis = 1)
dafc.head(5)
```

```python
dafc.shape
```

```python
dafc.info()
```

```python
dafc.isnull().sum()
```

```python
dafc = dafc.dropna(axis = 0)
```

```python
dafc.describe()
```

```python
#extracting the time element from row index 1

def get_time(element):
    return element[0:6]
```

```python
#change the data type to strings

dafc[1] = dafc[1].astype(object)
```

```python
#applying the get_time function to extract time

dafc['Time'] = dafc[1].apply(get_time)
dafc.head(10)
```

```python
column = ['Date', 'raw', 'Time']
dafc.columns = column
```

```python
#spliting the elements in raw column

dafc['raw'] = dafc['raw'].str.split("-")
```

```python
dafc['raw'] = dafc['raw'].str[1]
```

```python
#extracting the family member element of raw column

dafc['Member'] = dafc['raw'].str.split(':').str[0]
```

```python
#extracting the message element of raw column

dafc['Message'] = dafc['raw'].str.split(':').str[1]
dafc.head(10)
```

```python
dafc = dafc.drop(labels = 'raw', axis = 1)
dafc.head()
```
